```python
## Basic Syntax rules for Python 3.10
## 1. Python is case-sensitive.

name ="Deepak"
Name ="Rathor"

print(name) # Output: Deepak
print(Name) # Output: Rathor
```

```python
## Indentation is important in Python. It is used to define the scope of loops, functions, and other code blocks.
if name == "Deepak":
    print("Hello, Deepak!")
else:
    print("Hello, stranger!")
```

```python
## Line continuation in Python can be done using a backslash (\) or by enclosing the expression in parentheses.
total = 1 + 2 + 3 + \
        4 + 5
print(total) # Output: 15
```

```python
## Multiple statements on a single line can be separated by a semicolon (;).
x = 10; y = 20; print(x + y) # Output: 30
```

```python
## Sentences in Python
## Variables in Python can be assigned values of different types, and the type can change during execution.
my_variable = 10
print(type(my_variable)) # Output: 10 
my_variable = "Hello, World!"
print(type(my_variable)) # Output: Hello, World!
```
